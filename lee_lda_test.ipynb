```python
from NewsAnalyzer.PreProcesser import preprocesser

pp = preprocesser()

file_path = "data/2022-04-12 news data, test.csv"
df = pp.road_data(file_path)
df
```

```python
title = df["기사제목"]
```

```python
pp.preprocess(title)
```

```python
# 자연어인 기사제목을 단어 리스트로 변환
stemmed = []
for text in title:
    tt = pp.stemmer(text)
    stemmed.append(tt)

stemmed
```

```python
# 정수인코딩, 단어사전을 만들고 단어, 그리고 해당 단어가 문서에 몇 번 등장하였는지 (단어,문서에 단어등장횟수)
from gensim import corpora

dictionary = corpora.Dictionary(stemmed)
corpus = [dictionary.doc2bow(text) for text in stemmed]
```

```python
import gensim

NUM_TOPICS = len(title)
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=10)
topics = ldamodel.print_topics(num_words=4)
```

```python
for topic in topics:
    print(topic)
```
