```python
from tensorflow.keras.preprocessing.text import Tokenizer

def int_encoding(df):
  tokenizer = Tokenizer()
  tokenizer.fit_on_texts(df['과일'])
  word_dict = tokenizer.word_index

  return word_dict

int_encoding(title[1])
```

```python
def padding(df):
  tokenizer = Tokenizer()
  tokenizer.fit_on_texts(df['과일'])
  encoded = tokenizer.texts_to_sequences(df['과일']) # 뒤에 설정 넣을 수 있음(아래로)

  return encoded
```
